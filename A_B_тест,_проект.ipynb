# Введение

Компания, занимающаяся рекламными технологиями, недавно представила новую модель ставок за показы для своих онлайн-рекламных кампаний: Average Bidding, в качестве альтернативы существующей модели ставок: Maximum Bidding.


*Модель ставок за показы (в контексте онлайн-рекламы и цифрового маркетинга) относится к стратегии, используемой рекламодателями и рекламными платформами для определения того, сколько они готовы заплатить за показ своей рекламы конкретному пользователю при посещении им веб-сайта или приложения. Этот процесс является фундаментальным для покупки рекламного пространства в интернете и оптимизации рекламных кампаний.

# Предпосылка

Компания XYZ — клиент, проводящий ретаргетинговую рекламную кампанию  (кампанию, нацеленную на людей ранее взаимодействовавших с брендом) с использованием модели Maximum Bidding, решила протестировать эту новую модель ставок и хочет провести A/B тест, чтобы понять, работает ли Average Bidding лучше, чем Maximum Bidding.

*A/B тест — это метод эксперимента, в котором сравниваются две версии чего-либо (например, веб-сайта или рекламной кампании), чтобы понять, какая из них работает лучше. Пользователи делятся на две группы: одна видит версию A, другая — версию B. Во время A/B теста две разные версии показываются одновременно, а аудитория делится случайным образом. Затем собираются данные, и на их основе принимается решение о том, какой вариант эффективнее.

* Версия A — это исходная версия, которую уже используют или тестируют (её видит контрольная группа).
* Версия B — это новая или измененная версия, которую хотят протестировать (её видит тестовая группа).

# Дизайн эксперимента

**Метрики оценки эффективности (с учетом представленного датасета):**

* Click-Through Rate (CTR) = # of Website Clicks / # of Impressions
* Conversion Rate (CR) = # of Purchase / # of Website Clicks
* Cost Per Click (CPC) = Spend [USD] / # of Website Clicks
* Cost Per Acquisition (CPA) = Spend [USD] / # of Purchase

**Параметры эксперимента:**

* Цель эксперимента: Определить наиболее эффективную модель ставок за показы
* H(0): Разница в эффективности между средними ставками и максимальными ставками нет (статистически не значима)
* H(1): Разница в эффективности между средними ставками и максимальными ставками все же существует (статистически значима)
* Продолжительность эксперимента = 1 месяц
* α = 0.05
* β = 0.2
* Мощность = 1 - β = 0,8


**Определение размера групп:**

Для данного pet-проекта невозможно корректно определить размер группы, поскольку показатели метрик до проведения A/B-теста неизвестны. Это не позволяет установить обоснованный минимально ожидаемый эффект (MDE). Из-за отсутствия информации о MDE расчет мощности стат критерия также невозможен.

Минимальный размер группы рассчитывается отдельно для каждой метрики, как минимум потому что MDE может отличаться. Однако предположим, что данные в этом датасете собирались настолько большой группой пользователей, что для каждой метрики был достигнут минимальный размер группы, необходимый для статистически значимого анализа.

## Бутстрап

```python
from scipy import stats
import numpy as np

from joblib import Parallel, delayed
from tqdm import tqdm
```

```python
def bootstrap_sampling(control_data, test_data, n_iterations=10000, sample_size=None, parallel_threshold=10000, n_jobs=-1):
    """
    Выполняет бутстрап-семплирование для двух выборок, адаптивно выбирая между последовательным и параллельным выполнением.

    :param control_data: numpy array или pandas Series с данными контрольной группы
    :param test_data: numpy array или pandas Series с данными тестовой группы
    :param n_iterations: Количество итераций бутстрапа
    :param sample_size: Размер подвыборки (если None, используется размер меньшей выборки)
    :param parallel_threshold: Порог количества элементов, при превышении которого используется параллельное выполнение
    :param n_jobs: Количество параллельных задач для CPU (-1 означает использование всех доступных ядер)
    :return: Кортеж с результатами бутстрапа для контрольной и тестовой групп
    """
    control_data = np.array(control_data)
    test_data = np.array(test_data)
    
    if control_data.size == 0 or test_data.size == 0:
        raise ValueError("Входные данные не могут быть пустыми")
    
    if control_data.ndim != 1 or test_data.ndim != 1:
        raise ValueError("Входные данные должны быть одномерными")
    
    if np.isnan(control_data).any() or np.isnan(test_data).any():
        raise ValueError("Входные данные содержат NaN значения")
    
    if sample_size is None:
        sample_size = min(control_data.size, test_data.size)
    elif sample_size <= 0 or not isinstance(sample_size, int):
        raise ValueError("sample_size должен быть положительным целым числом")
    
    if n_iterations <= 0 or not isinstance(n_iterations, int):
        raise ValueError("n_iterations должно быть положительным целым числом")
    
    total_data_size = control_data.size + test_data.size
    if total_data_size > parallel_threshold:
        return _bootstrap_sampling_parallel(control_data, test_data, n_iterations, sample_size, n_jobs)
    else:
        return _bootstrap_sampling_sequential(control_data, test_data, n_iterations, sample_size)

def _bootstrap_iteration(control_data, test_data, sample_size):
    control_sample = np.random.choice(control_data, size=sample_size, replace=True)
    test_sample = np.random.choice(test_data, size=sample_size, replace=True)
    return np.mean(control_sample), np.mean(test_sample)

def _bootstrap_sampling_sequential(control_data, test_data, n_iterations, sample_size):
    control_means = np.zeros(n_iterations)
    test_means = np.zeros(n_iterations)
    
    for i in tqdm(range(n_iterations)):
        control_means[i], test_means[i] = _bootstrap_iteration(control_data, test_data, sample_size)
    
    return control_means, test_means

def _bootstrap_sampling_parallel(control_data, test_data, n_iterations, sample_size, n_jobs):
    results = Parallel(n_jobs=n_jobs)(
        delayed(_bootstrap_iteration)(control_data, test_data, sample_size)
        for _ in tqdm(range(n_iterations))
    )
    control_means, test_means = zip(*results)
    return np.array(control_means), np.array(test_means)

def analyze_results(control_means, test_means, alpha=0.05):
    """
    Проводит статистический анализ результатов бутстрапа.

    :param control_means: Массив средних значений для контрольной группы
    :param test_means: Массив средних значений для тестовой группы
    :param alpha: Уровень значимости для теста (по умолчанию 0.05)
    :return: Словарь с результатами анализа и интерпретацией
    """
    t_stat, p_value = stats.ttest_ind(control_means, test_means)
    
    results = {
        "p_value": p_value,
        "t_statistic": t_stat,
        "alpha": alpha
    }
    
    if p_value < alpha:
        results["conclusion"] = "Нулевая гипотеза отвергается."
        results["interpretation"] = (
            "Существует статистически значимое различие между контрольной и тестовой группами."
        )
    else:
        results["conclusion"] = "Нулевая гипотеза не отвергается."
        results["interpretation"] = (
            "Нет статистически значимого различия между контрольной и тестовой группами."
        )
    
    return results

def visualize_results(control_means, test_means):
    """
    Создает визуализацию результатов бутстрапа.
    
    :param control_means: Массив средних значений для контрольной группы
    :param test_means: Массив средних значений для тестовой группы
    :param results: Словарь с результатами анализа
    """
    plt.figure(figsize=(12, 8))

    # Гистограмма распределения средних значений
    plt.subplot(2, 1, 1)
    sns.histplot(control_means, kde=True, color='blue', alpha=0.5, label='Control')
    sns.histplot(test_means, kde=True, color='red', alpha=0.5, label='Test')
    plt.title('Распределение средних значений')
    plt.xlabel('Среднее значение')
    plt.ylabel('Частота')
    plt.legend()

    # График разницы средних значений
    plt.subplot(2, 1, 2)
    mean_differences = test_means - control_means
    sns.histplot(mean_differences, kde=True, color='green')
    plt.title('Распределение разницы средних значений')
    plt.xlabel('Разница средних значений')
    plt.ylabel('Частота')
    plt.legend()

    plt.tight_layout()
    plt.show()
```
